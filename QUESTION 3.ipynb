```python
def bubbleSort(my_list):
    for passnum in range(len(my_list)-1,0,-1):
        for i in range(passnum):
            if my_list[i]>my_list[i+1]:
                temp = my_list[i]
                my_list[i] = my_list[i+1]
                my_list[i+1] = temp

my_list = [29,13,22,37,52,49,46,71,56]
bubbleSort(my_list)
print(my_list)
```
